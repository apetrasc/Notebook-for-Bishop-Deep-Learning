# 10-0 Convolutional Neural Networks

最も簡便な機械学習モデルというものは、観測データが構造化されていない、つまりデータベクトル$\boldsymbol{x}=(x_1,..x_D)$について、個々の要素がどのように関連しているのかについて事前に何も知らないということを意味する。仮にこのような個別入力をランダムに順序を入れ替えるといった操作を行い、固定化された順序をすべての訓練データ・テストデータに適用した場合にはおおよそなんの影響も与えられないはずである。  
しかし、機械学習の多くの応用において、入力データそのもの同士が相関を持つ構造化データに対する適用を含んでいる。例えば、自然言語における単語は系列を成し、仮に言語を生成的自己回帰プロセスをもつとモデル化した場合には、これら単語同士は直前の言葉により強く影響され、最初の単語に対しては比較的弱い相関を持つと考えることができる。同様に、一枚の画像の中のピクセルは、2次元空間の中で入力変数同士が適切に定義された空間的相関をお互いに保持しており、近傍のピクセル同士が強い相関を持っている。  
我々はこれまで、特定のデータの構造に関する知識が、誤差関数に正則化項を追加したり、データをかさまししたり、モデルの設計を変更することによって組み込まれることを見てきた。こういった試みは、入力データに対する変換という観点における不変性や等価性といった特定の性質に着目する助けとなる。本章では畳み込みニューラルネットワーク(Convolutional Neural Network,CNN)と呼ばれる設計的アプローチに注目する。これは、パラメータ共有を行う局所的に接続された多層ネットワークと理解することができ、特に画像データに特化した普遍性や等価性を表現するために設計されたものである。
