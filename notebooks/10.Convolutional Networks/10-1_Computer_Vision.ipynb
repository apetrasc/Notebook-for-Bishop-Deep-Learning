# 10-1_Computer Vision

自動的に画像データを解析・解釈する技術はコンピュータビジョンの分野への注目を促し、機械学習の応用先として代表的な地位を占めるに至った。歴史的には、コンピュータビジョンは主に3次元投影の分野に基礎をおいていた。人の手によってつくられた特徴量が構築され、簡易的な学習アルゴリズムの入力として用いられてきた。しかし、これもまた、深層学習の発展によって大きな変化が起こった分野のうちの一つなのである。特にCNNの影響が大きいとされ、これは元来画像解析の分野で発展してきた歴史をもつものの、系列データの解析といった他分野に対しても応用されてきた。近年、一部の応用においてTransformersと呼ばれる代替的なモデルが対抗馬として人気を博してきている。
