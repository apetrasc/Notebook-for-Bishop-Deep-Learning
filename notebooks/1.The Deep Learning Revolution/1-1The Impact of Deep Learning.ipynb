# 1-1深層学習がもたらした衝撃
機械学習は多岐にわたる分野でその有効性を示しており、深層学習の一貫した枠組みを用いて多様な問題が解決されています。これは、専門的な技術を必要とする従来の方法とは対照的です。深層学習の適用可能性は非常に広く、計算技術が関わるほぼすべての分野で革新的な変化をもたらす可能性があります。


## 1-1-1 医療画像診断の例
機械学習を医療画像診断に使用する事例を紹介しています。メラノーマは皮膚がんの一種で危険ではありますが、早期発見ができれば治療ができます。しかし悪性か良性かを判断するのはやや難しいことが多いです。ルールベースでこれは悪性、これは良性というようにスクリプトを書いていてもルールが発散してしまうのです。そのような問題を解決するのが深層学習なのでした。ニューラルネットの重みとして知られる2500万ものパラメータを決定するために129,000もの画像が訓練に使用されました。しかしこれは比較的少数の部類に入ります。最初の訓練は、128万もの日常の風景を映した画像が訓練に使用されました。そのあと、ファインチューニングとして医療画像が用いられるのです。日々のありふれた物体の画像のデータセットから医療画像という専門性の高い画像へ特化していくことから、これは"transfer learning"と呼ばれます。これにより、専門家を超える検出精度をたたき出すことができるのでした。
