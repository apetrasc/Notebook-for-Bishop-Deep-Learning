# 1-1深層学習のインパクト
機械学習は多岐にわたる分野でその有効性を示しており、深層学習の一貫した枠組みを用いて多様な問題が解決されている。深層学習で作成されたアルゴリズムは伝統的なハンドメイドのものを置き換えつつある。
深層学習はデータから学ぶ方法として汎用的な枠組みを提供した。深層学習は人の脳の構造を模倣したニューラルネットワークを基盤としており、現在は特定の目的に特化した作成方法がなされており、人間の脳では到底できない作業を行うことができている。LLMとよばれる巨大な深層学習のシステムは、世界初の汎用人工知能と呼ばれるほどの性能を示している。


## 1-1-1 医療画像診断の例
機械学習を医療画像診断に使用する事例を紹介している。  
メラノーマ:  
〇皮膚がんの一種で危険ではありますが、早期発見ができれば治療可能  
△悪性か良性かを判断するのはやや難しい  
ルールベースでこれは悪性、これは良性というようにスクリプトを書いていてもルールが発散してしまう  
→深層学習  
ニューラルネットの重みとして知られる2500万ものパラメータを決定するために129,000もの画像（少ない部類に入る）が訓練に使用された。  
128万もの日常の風景を映した画像が訓練に使用され、そのあと、ファインチューニングとして医療画像が用いられる。→Transfer learning  
日々のありふれた物体の画像のデータセットから医療画像という専門性の高い画像へ特化していこと＝＞"transfer learning"  
専門家を超える検出精度をたたき出すことができた。  

## 1-1-2 タンパク質構造の同定
タンパク質は22種のアミノ酸の連鎖である。細胞内で合成されると3次元構造をなし、挙動や性質はその立体構造に大きく左右される。アミノ酸の連鎖からこの3Dモデルを計算する問題は生物学の基本的な問題で、深層学習到来まではほとんど進歩しなかった。  
3d構造は実験的に測定されるが、時間がかかり一部のタンパク質は困難であるとされる。対称的に、タンパク質の中で連鎖を成すアミノ酸は測定が容易。新薬の開発といった応用の目的で、アミノ酸の配列から3Dタンパク質の構造を同定することに大きな関心が払われている。既知のものに対しアミノ配列を入力、タンパク質の3D構造を出力として訓練される＝＞supervised learningと呼ばれる。


## 1-1-3 画像合成
ニューラルネットワークは入力、出力の関係だけにとどまらない。サンプル画像の集合だけから、それらに似た画像を新たに創造するネットワークの作成にも使用可能（=unsupervised learning）  
新たな画像を創造できることから生成モデル(generative model)と呼ばれるが、これは統計学的な性質を共有している。プロンプトとして知られる既知の文字列から画像が生成されるためテキスト入力の意味を反映している。

## 1-1-4 LLM(大規模言語モデル)
機械学習におけるもっとも重要な進歩の一つに、自然言語やソースコードのような時系列データを処理するモデルの進歩がある。LLMは深層学習を用い言語の意味論的な性質をとらえる豊かな内部表現を構築している。自己回帰モデルは言語を生成できるので、生成AIの一種である。そういったモデルは、ある単語の次にどの単語が続くのかを系列で学習する。このプロセスを繰り返し、長い文章を生成できるのだ。文を止めるピリオドの存在も認識しており、生成がとまった場合には人間は単語を追加し、さらなる単語生成をトリガーできる。→会話できる  
そういったモデルは、ランダムに選ばれた単語の羅列のペアを抽出した文字列を入力とする。これはself-supervised learningと呼ばれ、inputからoutputへの関数を学習するが、ラベルがつけられた出力は入力から自動的に取得される。いかなる情報源も利用可能なのでテキストデータは膨大となる。GPT-4と呼ばれる言語生成のモデルは世界初の汎用人工知能（AGI）と呼ばれるほどの性能を示した。
